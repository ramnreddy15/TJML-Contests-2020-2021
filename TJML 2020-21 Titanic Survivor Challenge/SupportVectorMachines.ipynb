```
df = pd.read_csv("/content/drive/My Drive/Support-Vector-Machines-from-Scratch/data/training.csv")
train_data_df = pd.read_csv("/content/drive/My Drive/Support-Vector-Machines-from-Scratch/data/training.csv") 
test_data_df = pd.read_csv("/content/drive/My Drive/Support-Vector-Machines-from-Scratch/data/testing.csv")

train_data_df
```

```
train_data_df['Sex'] = train_data_df['Sex'].map({'male': 0, 'female': 1}).astype(int)
train_data_df['Age'] = train_data_df.Age.round()
test_data_df['Sex'] = test_data_df['Sex'].map({'male': 0, 'female': 1}).astype(int)
test_data_df['Age'] = test_data_df.Age.round()
#train_data_df['Age'] = train_data_df.groupby(['Survived','Embarked'])['Age'].apply(lambda x: x.fillna(x.mean()))
#test_data_df['Age'] = test_data_df.groupby(['Sex','Embarked'])['Age'].apply(lambda x: x.fillna(x.mean()))

#test_data_df['Age'] = test_data_df['Age'].fillna(df.groupby('Embarked')['Age'].transform('mean'))
#train_data_df['Age'] = train_data_df['Age'].fillna(df.groupby(['Survived','Embarked'])['Age'].transform('mean'))
test_data_df['Age'] = test_data_df['Age'].fillna(test_data_df['Age'].median())
train_data_df['Age'] = train_data_df['Age'].fillna(train_data_df['Age'].median())
train_data_df['Embarked'] = train_data_df['Embarked'].fillna(df['Embarked'].value_counts().index[0])
train_data_df['Cabin'] = train_data_df['Cabin'].fillna('Unknown')
test_data_df['Embarked'] = test_data_df['Embarked'].fillna(df['Embarked'].value_counts().index[0])
test_data_df['Cabin'] = test_data_df['Cabin'].fillna('Unknown')
train_data_df
```

```
train_data_df = train_data_df[train_data_df.Age.notnull()][['Age','Sex',  'Pclass', 'Parch','Fare', 'SibSp','Survived','Embarked','Cabin']]

from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()

train_data_df['Cabin'] = LE.fit_transform(train_data_df['Cabin'])
test_data_df['Cabin'] = LE.fit_transform(test_data_df['Cabin'])
train_data_df['Embarked'] = LE.fit_transform(train_data_df['Embarked'])
test_data_df['Embarked'] = LE.fit_transform(test_data_df['Embarked'])
```

```
test_data_df.Age.isnull().sum()
X = train_data_df[['Sex',    'Age', 'Pclass','SibSp', 'Fare']]
x_test = test_data_df[[ 'Sex',   'Age', 'Pclass','SibSp', 'Fare']]

Y = train_data_df['Survived']
x_test
```

```
from sklearn.metrics import accuracy_score

from sklearn.svm import SVC

classifier = SVC()

classifier.fit(X, Y)



y_pred= classifier.predict(x_test)
classifier.score(X,Y)
```
