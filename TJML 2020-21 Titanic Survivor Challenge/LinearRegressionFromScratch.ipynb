```
def changeLine(epochs, learningRate, m, b, dataset):
  for n in range(epochs):
    m,b = doMath(learningRate,m,b,dataset)
  return m,b
```

```
def distanceFromPointLines(m,b,x1,y1):
  return y1 - (m*x1 + b)
```
